## 1. Setup Environment

```python
# Install system dependencies
!apt-get update -qq
!apt-get install -y ffmpeg libsm6 libxext6 libxrender-dev libgomp1

import os
import zipfile

PROJECT_ZIP = "facefusion_project.zip"

if os.path.exists(PROJECT_ZIP):
    print(f"Found {PROJECT_ZIP}, unpacking...")
    
    if os.path.exists('facefusion_app'):
        !rm -rf facefusion_app
    
    os.makedirs('facefusion_app', exist_ok=True)
    with zipfile.ZipFile(PROJECT_ZIP, 'r') as zip_ref:
        zip_ref.extractall('facefusion_app')
    
    print("Unpacked successfully!")
else:
    print(f"⚠️ {PROJECT_ZIP} not found! Please upload it to the Files sidebar.")
```
